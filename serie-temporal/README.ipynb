# 5. Como Executar (passo a passo, caso seja necessário)
5.1 Estrutura de arquivos necessária
projeto-int/
├── main.py

├── 01_series_temporais.py

├── 02_previsao.py

├── PI3_SeriesTemporais_MannKendall.ipynb

└── dados/
  
    └── comparativo_mensal.csv

 O CSV precisa estar dentro de uma pasta chamada exatamente 'dados', no mesmo nível dos arquivos .py.


# 5.2 Instalar dependências (uma vez só)
Abra o Prompt de Comando (cmd) na pasta do projeto e rode:
pip install streamlit pandas numpy plotly scipy statsmodels pymannkendall
Se aparecer 'Requirement already satisfied' para tudo, já está instalado — pode seguir.

# 5.3 Executar o Streamlit (dashboard interativo)
Se o comando 'streamlit' não for reconhecido, use:
python -m streamlit run main.py
Ou, para abrir só a página de séries temporais isoladamente:
python -m streamlit run 01_series_temporais.py
Na primeira execução, o Streamlit pode perguntar um e-mail de cadastro. Basta apertar Enter sem digitar nada para pular.
Depois disso, o navegador abre automaticamente em http://localhost:8501. Se não abrir, copie esse endereço manualmente para o navegador.
Para parar o servidor, volte ao terminal e pressione Ctrl+C.


# 5.4 Executar o Notebook (Jupyter)
Opção A — Jupyter Notebook clássico:
pip install notebook
jupyter notebook
Isso abre o navegador com a lista de arquivos. Clique em PI3_SeriesTemporais_MannKendall.ipynb e depois em 'Run All' (ou Cell > Run All) para executar tudo.
Opção B — VS Code: abra o arquivo .ipynb diretamente no VS Code (com a extensão Jupyter instalada) e clique em 'Run All' no topo do notebook.


# 5.5 Problemas comuns e soluções

'streamlit' não é reconhecido	Use 'python -m streamlit run ...' em vez de 'streamlit run ...'

FileNotFoundError: dados/comparativo_mensal.csv	Confirme que a pasta 'dados' existe no mesmo diretório do script, com o CSV dentro

ModuleNotFoundError: No module named 'X'	Rode: pip install X  (ou pip install -r requirements.txt se o arquivo existir)

Gráfico não aparece / célula trava no Jupyter	Certifique-se de executar as células em ordem (Run All), pois células posteriores dependem de variáveis criadas antes

Pede e-mail e trava	É o cadastro opcional do Streamlit na primeira vez — aperte Enter em branco para pular
